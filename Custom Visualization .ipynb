```python
#RESAMPLING AND ROLLING AVERAGE PLOTS
```

```python
ppc_data.head()
```

```python
#lets look at the cost data from a ppc dataset
ppc_data['Cost'].plot(figsize=(15,5))
```

```python
##Weekly sampling
ppc_weekly = ppc_data['Cost'].resample('W').mean()
ppc_weekly.head()
```

```python
ppc_weekly.plot(figsize=(15,5))
```

```python
#Rolling AVERAGE
ppc_rolling_7d = ppc_data['Cost'].rolling(window=7, center=True
                                                   ).mean()

ppc_rolling_7d.head(10)
```

```python
ppc_rolling_7d.plot(figsize=(15,5))
```

```python
#Plot Weekly PPC CPC Comparison
#i will plot three different CPC data points, Daily, Resampled Weekly and Rolling Weekly then overlay and compare the output.
import numpy as np
import matplotlib
```

```python
fig = plt.figure(figsize=(15,5))

ax = fig.add_subplot(111)

figsize = (15,8)

max_value = ppc_rolling_7d.max()
max_value_index = ppc_rolling_7d.idxmax()

min_value = ppc_rolling_7d.min()
min_value_index = ppc_rolling_7d.idxmin()

ppc_data['Cost'].plot(figsize=figsize,
                      alpha=.5,
                      marker='.',
                      linestyle='-',
                      linewidth=0.5,
                      label='Daily'
                     )

ppc_weekly.plot(figsize=figsize,
                marker='x',
                markersize=8,
                linestyle='-',
                label='Weekly Mean Resample'
               )

ppc_rolling_7d.plot(figsize=figsize,
                    marker='o',
                    linestyle='-',
                    label='7-d Rolling Mean'
                   )
#Create dyanmic annotations for max Values
ax.annotate('Highest CPC: {}'.format(max_value),
            xy=(max_value_index,max_value),
            xytext=(max_value_index,max_value),
            arrowprops=dict(facecolor='cyan',
                            shrink=0.05,
                            lw=1,
                            ec='magenta',
                            zorder=1))

ax.annotate('lowest CPC: {}'.format(min_value),
            xy=(min_value_index,min_value),
            xytext=(min_value_index,min_value),
            arrowprops=dict(facecolor='cyan',
                            shrink=0.0,
                            lw=1,
                            ec='magenta',
                            zorder=1))

ax.set_xlabel('Date')
ax.set_ylabel('Cost Per Click (CPC)')

#Global plot Settings
plt.title('COMPARE; Daily, Weekly Mean, 7-d Rolling Mean')
fig.legend()

plt.show()
```

```python
#Create dyanmic annotations for max Values
ax.annotate('Highest CPC: {}'.format(max_value.round()),
            xy=(max_value_index,max_value),
            xytext=(max_value_index,max_value),
            arrowprops=dict(facecolor='cyan',
                            shrink=0.5,
                            lw=1,
                            ec='magenta',
                            zorder=1))

ax.annotate('lowest CPC: {}'.format(min_value.round())
            xy=(min_value_index,min_value),
            xytext=(min_value_index,min_value),
            arrowprops=dict(facecolor='cyan',
                            shrink=0.5,
                            lw=1,
                            ec='magenta',
                            zorder=1))
```
